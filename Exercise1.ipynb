```python
import pandas as pd
```

```python
df_SPX = pd.read_csv('https://s3.eu-west-1.amazonaws.com/neueda.conygre.com/pydata/SPX.csv', index_col='Date', parse_dates=True)
```

```python
df_SPX.head()
```

```python
df_SPX['Price'] = pd.to_numeric(df_SPX['Price'].str.replace(',', ''))
df_SPX['Open'] = pd.to_numeric(df_SPX['Open'].str.replace(',', ''))
df_SPX['High'] = pd.to_numeric(df_SPX['High'].str.replace(',', ''))
df_SPX['Low'] = pd.to_numeric(df_SPX['Low'].str.replace(',', ''))
df_SPX['Change %'] = pd.to_numeric(df_SPX['Change %'].str.replace('%', ''))
```

```python
df_SPX.sort_values('Date', inplace=True)
```

```python
import matplotlib.pyplot as mtplot

col = ['High' , 'Low']


df_SPX.query('Date >= 2015 & Date <= 2016')[col].plot(figsize = (18,9))
mtplot.show()

```
